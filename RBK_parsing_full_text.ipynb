```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
import pandas as pd
import pickle
```

```python
df_test = pd.read_csv("dataset/test_dataset_test.csv")
```

```python
df_test.sample()
```

```python
### парсинг новостей
all_news_test = []
for id_news in df_test['document_id'].tolist():
    try:
        text_news = ''
        driver = webdriver.Chrome(executable_path="/home/eugene/Desktop/RBK_competetion/utils/chromedriver")
        driver.get(url=f"https://www.rbc.ru/rbcfreenews/{id_news[0:24]}")
        time.sleep(20)
        for i in driver.find_elements(By.TAG_NAME, "p"):
            text_news += i.text
        all_news_test.append(text_news)
    except Exception as ex:
        print(ex)
        append(None)
    finally:
        driver.close()
        driver.quit()
```

```python
with open('utils/full_news_rbk_test.pickle', 'wb') as handle:
    pickle.dump(all_news_test, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open('utils/full_news_rbk_test.pickle', 'rb') as handle:
    all_news_test = pickle.load(handle)
```

```python
df_test['full_text'] = all_news_test
```

```python
df_test.query("full_text==''")['document_id'] # статьи не загружены
```

```python
# незагруженные статьи добавляем вручную
with open('utils/add_full_news_rbk.pickle', 'rb') as handle:
    add_full_news_rbk = pickle.load(handle)
```
